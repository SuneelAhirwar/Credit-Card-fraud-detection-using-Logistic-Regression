### We will load the dependencies for model building

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

##### after loading the dependencies we will use the dataset. 

```python
#load the credit card datasets
card_data = pd.read_csv('creditcard.csv')
```

```python
# Now we will classify the class that which class has more values
card_data['Class'].value_counts()
```

```python
#separating the data for analysis
legit = card_data[card_data.Class ==0]
fraud = card_data[card_data.Class==1]
```

```python
## Compare the values for both transaction
card_data.groupby('Class').mean()
```

```python
# we will use sample data of normal transactions
legit_sample = legit.sample(n=492)
```

```python
## Cancating the  legit sample and fraud transactions
new_data = pd.concat([legit_sample,fraud],axis=0)
```

#### Now we will split the data into independent and dependent features


```python
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into Training and Testing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

#### We will use the Logistic Algorithm for classification model building. 

```python
model = LogisticRegression()
```

```python
#Train the logistic regression model with Training data
model.fit(X_train,Y_train)
```

```python
## We will check the accuracy on training data
X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
test_data_accuracy
```

```python
print('Accuracy on Test data: ', test_data_accuracy)
```
